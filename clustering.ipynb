## KMeansClustering


K-Means Clustering is a form of unsupervised learning. Unsupervised learning is the process by which the algorithm sorts the data into k groups, where k is determined by the user. The process of the algorithm is as follows:

    -clustering algorithm creates k centroids by randomly picking a point from the existing data
    -algorithm assigns every other point to the nearest centroid
    -algorithm re-defines the centroid by taking the average location of the points assigned to it
    -algorithm re-evaluates the distance between every data point and the newly defined centroids
    -algorithm re-assigns the datapoints based on which centroid is closest
    -algorithm re-defines centroids by averaging the location of every point now assigned to it
    
This process takes place several times and eventually the centroids reach an equilibrium where their position no longer changes (or changes relatively little) after each re-evaluation.

Let's take a look at how you would code something like this.

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn import datasets

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')
cmap = 'tab10'

dummy_dat = pd.read_csv("xclara.txt", header=0, index_col=0)
dummy_dat.reset_index(inplace=True)
dummy_dat.drop('index', axis=1, inplace=True)
```

```python
# initiate model and fit to the dummy data
model = KMeans(n_clusters=3, random_state=42)
model.fit(dummy_dat)
```

```python
# look @ the data in raw form

plt.scatter(dummy_dat['V1'], dummy_dat['V2']);
```

```python
# look at the centroids created by kmeans algorithm

plt.scatter(dummy_dat['V1'], dummy_dat['V2'])
for i in range(len(model.cluster_centers_)):
    plt.scatter(model.cluster_centers_[i][0],
                model.cluster_centers_[i][1]);
```

## Conclusion

The KMeans algorithm illustrates how well it can find the center of these clusters. Please note that these clusters were already defined as such in the data, and KMeans will not perform this perfectly in the real world. Put simply, real world data is not naturally sorted into conventient clusters, and KMeans will only get you so far.
